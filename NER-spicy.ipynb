```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
from pprint import pprint
import docx
nlp = en_core_web_sm.load()
```

```python
# data = pd.re
try:
    doc = docx.Document('./zycus_assignment/Training_data/6683127-House-Rental-Contract-GERALDINE-GALINATO-v2-Page-1.pdf.docx')  # Creating word reader object.
    data = ""
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
        data = '\n'.join(fullText)

    print(data)

except IOError:
    print('There was an error opening the file!')
```

```python
doc = nlp(data)
```

```python
preprocessed_doc = ""
for token in doc:
#     print(token.text)
    preprocessed_doc = preprocessed_doc + "   " + token.text

print(preprocessed_doc)
```

```python

text_extra_info = [X.text for X in doc.ents]
print(text_extra_info)
```

```python
import re
import csv

fields = ['left', 'center', 'right']
file_name = "text_aggre_extraction.csv"

with open(file_name, 'w') as csvfile: 
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(fields)
    
    for word in text_extra_info:
        try:
            r1 = re.search("(?:[a-zA-Z'-]+[^a-zA-Z'-]+){0,5}"+word+"(?:[^a-zA-Z'-]+[a-zA-Z'-]+){0,5}", preprocessed_doc, re.IGNORECASE)
            sen = r1.group()

            sen = sen.partition(word)
            left_context = sen[0].split()
            right_context = sen[2].split()
            row = left_context,word,right_context
            print(row) 
            csvwriter.writerow([left_context,word,right_context]) 
            csvfile.flush()
        
        except:
            print("pass")
            pass
    
```

```python
def clean_dict(dict,idx):
    dict[idx] = list(set(dict[idx]))
    if(idx == min(dict[idx])):
        return
    else:
        dict[min(dict[idx])] = dict[idx]
        del dict[idx]
        return
```

```python
Person = []

for x in doc.ents:
    if(x.label_ == 'PERSON'):
        if(len(x.text)<4):
            continue
        Person.append(x.text)

a = list(set(Person))

word2id = {}

for i in a:
    print(i,data.find(i))
    word2id[i] = data.find(i)
    
pprint(word2id)



closeness = [ [0]*len(a) for i in range(len(a))]





id2word= {}

for i in a:
    id2word[data.find(i)] = i
pprint(id2word)



for i in range(len(a)):
    for j in range(len(a)):
        if(i==j):
            closeness[i][j] =1000000
            continue
        closeness[i][j] = abs(data.find(a[i]) - data.find(a[j]))
pprint(closeness)


clusters = {}
# clusters[data.find(a[0])] = [data.find(a[0])]

for i in range(len(a)):
    if(word2id[a[closeness[i].index(min(closeness[i]))]] not in clusters.keys()):
        clusters[word2id[a[closeness[i].index(min(closeness[i]))]]] = [word2id[a[closeness[i].index(min(closeness[i]))]]]
    clusters[word2id[a[closeness[i].index(min(closeness[i]))]]].append(data.find(a[i]))
    id = word2id[a[closeness[i].index(min(closeness[i]))]]
    clean_dict(clusters,word2id[a[closeness[i].index(min(closeness[i]))]])

pprint(clusters)

for idx,i in enumerate(clusters.values()):
    print("Cluster {} : ".format(idx+1))
    for j in i:
        print(id2word[j],sep=", ")
    print()


        
```
